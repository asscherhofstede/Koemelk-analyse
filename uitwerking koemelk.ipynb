# Data prep

gezien dat wij als groep werken hebben wij het werk verdeeld in verschillende data prep stukjes. Dit zorgde ervoor dat ik de opdracht kreeg : het gemiddelde van elke sample en een gespecificeerd range van 1000 tot 3000. 
 

```python
# imports toevoegen
import os
import spc
import pandas as pd
import numpy as np

# lege dataframe aanmaken
df = []

# een list van een directory maken
listdir = os.listdir('Spectra')

# loopen in de directory s is het bestandsnaam
for s in listdir:

    # path variabel van de absolute path
    path = os.path.join("D:\\ai_semesterwindesheim\\koemelk\\Spectra", s)
    
    f = spc.File(path)  # read file
    file = f.data_txt()  # output data

    # df maken van de data
    df_in_one_go=pd.DataFrame(np.array([f.sub[0].y,f.sub[1].y,f.sub[2].y]),columns=f.x)

    # column toevoegen met samplename en s is het bestandsnaam die als value wordt toegevoegd
    df_in_one_go['samplename'] = s

    # voeg hier de data toe aan de lege dataframe, na de volgende loop zal die constant worden aangevuld
    df.append(df_in_one_go)

# de dataframes aan elkaar koppelen, en van een serie naar een dataframe brengen.
df_mean = pd.concat(df)
df_mean = pd.DataFrame(df_mean)

```

In het volgende code blokje maak ik van mijn samples dus allemaal gemiddeldes. nadat ik dat heb gedaan was het niet helemaal gelukt om code matig een nieuwe index toetevoegen. Doordat ik een index heb gemaakt in excel was dit wel gelukt. 

nadat ik dat heb gedaan voeg ik ze weer samen meegenomen dat ze gelijk zijn.

```python
# gemiddelde van alle samples maken
df_average = df_mean.groupby(np.arange(len(df_mean))//3).mean()

#import een hulp excel
df_excel = pd.read_excel('index.xlsx')
df_excel = df_excel.set_index('index')

#beide dataframes aan elkaar toevoegen
df = pd.concat([df_excel,df_average], axis=1)
```

nu moeten we het uitlezen van de featuresm dit is te vinden in refrence values xlsx

```python
# uitlezen van excel
df_refrence = pd.read_excel('Reference values.xlsx')
```

nadat deze is uitgelezen zullen we de gemiddelde samples mergen met het net uitgelezen dataframe. Dit zal op basis van de index gaan

```python
df_combined = pd.merge(df, df_refrence, left_on=df.samplename, right_on=df_refrence.Sample)
```

# Modeling

tijdens het modeleren gaan we met meerdere eigenschappen te maken hebben, wij gaan afzonderlijk van elkaar Fat, Protein en lactose controleren. hierdoor krijgen we drie modellen.

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split

# het splitsen van de data voor het model
X_train_all, X_test_all, y_train_all, y_test_all = train_test_split(df_combined[['Fat (%m/m)','Protein (%m/m)','Lactose (%m/m)']], df_combined.iloc[:,158:678], test_size=0.2, random_state=11)
print(X_train_all.shape, X_test_all.shape, y_train_all.shape, y_test_all.shape)

```

we werken hier met een afgesproken standaard model zodat elke notebook dezelfde modellen gebruikt en hierdoor geen verwarringen ontstaan.

```python
from sklearn.linear_model import MultiTaskLassoCV
from sklearn.metrics import r2_score

# model uitvoeren
reg_all = MultiTaskLassoCV(cv=5, random_state=11).fit(X_train_all, y_train_all)

# score van het model
r2_score(y_test_all, reg_all.predict(X_test_all))
```
